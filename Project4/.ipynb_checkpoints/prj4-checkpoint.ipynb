```python
df = df.iloc[:, 0:4]
dec_df = df.loc[df['month']==12]
dec_df = dec_df.drop(['month'], axis=1)
```

```python
dec_df.head()
```

```python
dec_df['sourceid'] = dec_df['sourceid'].astype(str)
dec_df['dstid'] = dec_df['dstid'].astype(str)
tuple_df = [tuple(col) for col in dec_df.values]
g = Graph.TupleList(tuple_df, directed = False, edge_attrs = ['mean_travel_time'])
```

```python
with open(directory+'/los_angeles_censustracts.json', 'r') as f:
    data_json = json.load(f)
geo_df = data_json['features']
```

```python
location = {}
for item in geo_df:
    data={}
    coor_list = item['geometry']['coordinates'][0][0]
    mean_coor = np.mean(np.asarray(coor_list), axis=0)
    movement_id = item['properties']['MOVEMENT_ID']
    
    data['coor_list'] = coor_list
    data['mean_coor'] = mean_coor
    data['display_name'] = item['properties']['DISPLAY_NAME']
    
    location[movement_id] = data
```
